```python
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.rand(10, 10)
y = np.random.rand(10, 10)
```

```python
def jaccard_precision_recall(gt, pred, threshold=0.5):
    assert gt.max() <= 1 and gt.min() >= 0
    assert pred.max() <= 1 and pred.min() >= 0
    assert threshold <= 1 and threshold >= 0
    assert gt.shape[0] == pred.shape[0]
    assert gt.shape[1] == pred.shape[1]

    gt = (gt >= threshold)
    pred = (pred >= threshold)
    
    inter = (gt & pred).sum()
    gt_area = gt.sum()
    pred_area = pred.sum()
    union = gt_area + pred_area - inter
    
    if union > 0:
        jaccard = inter / union
    else:
        jaccard = 0
    if pred_area > 0:
        precision = inter / pred_area
    else:
        precision = 0

    return jaccard, precision

def jaccard_precision_curve(gt, pred, threshold_num=1000):
    threshold_list = []
    jaccard_list = []
    precision_list = []
    for th in np.linspace(0, 1, threshold_num):
        threshold_list.append(th)
        j, p = jaccard_precision_recall(gt, pred, th)
        jaccard_list.append(j)
        precision_list.append(p)

    return threshold_list, jaccard_list, precision_list

```

```python
x_list = []
jaccard_list = []
precison_list = []
for i in range(1000):
    th = i / 1000
    j, p = jaccard_precision_recall(x, y, th)
    jaccard_list.append(j)
    precison_list.append(p)
    x_list.append(th)
```

```python
metrics.auc(x_list, jaccard_list)
```

```python
plt.plot(x_list, jaccard_list)
plt.plot([0, 1], [1, 0])
```

```python
metrics.auc(x_list, precison_list)
```

```python
plt.plot(x_list, precison_list)
plt.plot([0, 1], [1, 0])
```
